```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('voice.csv')
```

```python
df.head(5)
```

```python
df['label'].replace(to_replace = ['male', 'female'], value = [0, 1], inplace = True)
```

```python
df.head(5)
```

```python
y = df['label']
```

```python
y
```

```python
tf = df.drop(columns = ['label'], axis = 1)
```

```python
tf.head(5)
```

```python
tf.head(20).describe()
```

```python
# tf1 = (tf-np.mean(tf))/ (np.std(tf))
```

```python
from scipy.stats import zscore
```

```python
tf1 = tf.apply(zscore) 
```

```python
tf1.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(tf1, y, test_size= 0.2, random_state= 43)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
```

```python
classifier.get_params()
```

```python
classifier.fit(X_train, y_train)
```

```python
predictions = classifier.predict(X_test)
```

```python
predictions
```

```python
y_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
import seaborn as sns
```

```python
cm = confusion_matrix(predictions, y_test)
```

```python
cm
```

```python
report = classification_report(y_test, predictions,digits = 4)
print(report)
```

```python
import matplotlib.pyplot as plt

sns.heatmap(cm, annot=True, fmt='', cmap='Blues',xticklabels = ['male', 'female'],yticklabels = ['male', 'female'] )

# Add labels, title, and adjust layout
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.tight_layout()

# Display the heatmap
plt.show()
```

```python
tf2 = pd.concat([tf1, y], axis = 1)
```

```python
correlation_matrix = tf2.corr(method='pearson')
```

```python
correlation_matrix
```

```python
plt.figure(figsize=(18, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix - Voice Dataset')

# %matplotlib QT
# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
